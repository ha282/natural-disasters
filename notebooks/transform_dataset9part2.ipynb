# Exploratory Data Analysis of Natural Disasters Dataset

```python
import pandas as pd
```

```python
# Create a new DataFrame for films
natural_disasters_df = pd.read_csv("../data/raw/uncleaned-dataset9.csv")

# Display info about the dataframe
natural_disasters_df.head(10)
```

# Null Handling

```python
# Check for null values
natural_disasters_df.isnull().sum()
```

```python
null_percent = (natural_disasters_df.isnull().mean() * 100)
print(null_percent)
```

```python
df2 = natural_disasters_df.dropna(axis=1, thresh=10000).copy()
df2.isnull().values.any()

print("Before dropping col:", natural_disasters_df.shape)
print("After dropping col:", df2.shape)



```

```python
df3 = df2.dropna()
df3.isnull().values.any()

print("Before dropping records:", df2.shape)
print("After dropping record:", df3.shape)
```
