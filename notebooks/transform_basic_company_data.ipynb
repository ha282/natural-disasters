# Null Handling

```python
# Check for null values
basic_company_data_df.isnull().sum()
```

```python
null_percent = (basic_company_data_df.isnull().mean() * 100)
print(null_percent)
```

```python
df2 = basic_company_data_df.dropna(axis=1, thresh=10000).copy()
df2.isnull().values.any()


print("Before dropping col:", basic_company_data_df.shape)
print("After dropping col:", df2.shape)

df3 = df2.dropna()
df3.isnull().values.any()

print("Before dropping records:", df2.shape)
print("After dropping record:", df3.shape)
```
